```python
import numpy as np
import pandas as pd
import os
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

extdatapath = '../Datasets/TaxiBJ/TAXIBJ.ext'
extdata = pd.read_csv(extdatapath)
extdata.head()
```

```python
import datetime

date_format = '%Y-%m-%dT%H:%M:%SZ'
datetime.datetime(2013, 2, 1).weekday()
# add a day column: 0 = 'Monday', 1 = 'Tuesday', ... 6 = 'Sunday'
extdata.insert(2, 'day', pd.to_datetime(extdata['time'], format=date_format).dt.dayofweek)
extdata.head()
```

```python
# normalize temperatures to [-1, 1]
extdata['Temperature'] = 2* (extdata['Temperature'] - extdata['Temperature'].min())/(extdata['Temperature'].max() - extdata['Temperature'].min()) - 1
# normalize temperatures to [0, 1]
extdata['WindSpeed'] = (extdata['WindSpeed'] - extdata['WindSpeed'].min())/(extdata['WindSpeed'].max()-extdata['WindSpeed'].min())
extdata.head()
```
